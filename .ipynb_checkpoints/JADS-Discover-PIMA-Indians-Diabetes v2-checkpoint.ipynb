# Pima Indians Diabetes dataset

Predict the onset of diabetes based on diagnostic measures.

## Data description
This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective is to predict based on diagnostic measurements whether a patient has diabetes.

Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

- Pregnancies: Number of times pregnant
- Glucose: Plasma glucose concentration a 2 hours in an oral glucose tolerance test
- BloodPressure: Diastolic blood pressure (mm Hg)
- SkinThickness: Triceps skin fold thickness (mm)
- Insulin: 2-Hour serum insulin (mu U/ml)
- BMI: Body mass index (weight in kg/(height in m)^2)
- DiabetesPedigreeFunction: Diabetes pedigree function
- Age: Age (years)
- Outcome: Class variable (0 or 1)

````python
def expl_df(df_input, name):
    
    """ Prints analysis of dataframe 'df_input' with name 'name' and returns lists of numeric, categorical, complete,
        incomplete features, features with many zeros and features with high skew > 0.75. """

    print('\n\033[1mDataframe analysis for dataframe:  ',name,'\033[0m\n')

    # Print dimensions of dataframe
    num_rows = df_input.shape[0]
    num_cols = df_input.shape[1]
    print('Dataframe',name,'has', num_rows, 'rows and', num_cols,'columns\n')

    # Check which features are numeric and which features are categorical 
    num_features = list(df_input.select_dtypes(include=['int64','float64']).columns)
    cat_features = list(df_input.select_dtypes(include=['object', 'category']).columns)
    print('Dataframe',name,'has', len(num_features) , 'numeric features and', len(cat_features),'categorical features\n')

    # Check which features have missing values and might need attention and print number of features
    count_values_feature=df_input.count()

    complete_features = list(count_values_feature[count_values_feature == num_rows].index)
    incomplete_features = list(count_values_feature[count_values_feature < num_rows].index)
    print('Dataframe',name,'has',len(complete_features), 'complete features and', len(incomplete_features),'incomplete features\n')

    # Calculate skew for each numeric feature and select features with skew>0.75 
    calc_skew_feature = df_input[num_features].skew()
    skew_features = list(calc_skew_feature[calc_skew_feature > 0.75].index)
    print('Dataframe',name,'has',len(skew_features), 'numeric features with a skew > 0.75\n')
    
    return num_features, cat_features, incomplete_features, complete_features,skew_features
````

````python
#
# 1D histogram plots for all numeric features in dataframe without features in list
# 'list_of_features_to_be_dropped' sorted by feature name 
#

def plt_1D_histograms_for_dataframe_numeric_features(dataframe, list_of_features_to_be_dropped):

    """ Plot 1D histogram plots for all numerical features in pandas dataframe sorted by feature name """

    # Create list with Numerical Features of dataframe
    numeric_features = dataframe.select_dtypes(include=['int64','float64']).columns

    # Drop features as per list_of_features_to_be_dropped and convert to list
    if len(list_of_features_to_be_dropped) > 0:
        numeric_features = list(numeric_features.drop(list_of_features_to_be_dropped))
    
    # Calculate number of plots per column
    num_plt_cols = len(numeric_features)
    first_num_plt_cols = num_plt_cols - int(num_plt_cols/2)

    # Create subplots in 2 columns
    fig, axes = plt.subplots(first_num_plt_cols, 2, figsize=(16,first_num_plt_cols))

    # Create subplots in first column
    i=0
    for column in numeric_features[0:first_num_plt_cols]:
        dataframe[column].dropna(inplace=True)
        bins = np.histogram(dataframe[column], bins=100)[0]
        bins = np.expand_dims(bins, axis=0)
        axes[i,0].imshow(bins, cmap='YlOrRd', aspect='auto')
        axes[i,0].get_xaxis().set_visible(False)
        axes[i,0].get_yaxis().set_visible(False)
        axes[i,0].set_title(str(column), loc='left')
        i = i+1

    # Create subplots in second column
    i=0
    for column in numeric_features[first_num_plt_cols:num_plt_cols]:
        dataframe[column].dropna(inplace=True)
        bins = np.histogram(dataframe[column], bins=100)[0]
        bins = np.expand_dims(bins, axis=0)
        axes[i,1].imshow(bins, cmap='YlOrRd', aspect='auto')
        axes[i,1].get_xaxis().set_visible(False)
        axes[i,1].get_yaxis().set_visible(False)
        axes[i,1].set_title(str(column), loc='left')
        i = i+1

    plt.tight_layout()
    plt.show()
    return
````

# Modeling

````python
#
# Select which feature engineering steps you would like to apply. 
#   1A. Remove individuals with zero measurements for SkinThickness or Insulin
#   1B. Impute all zeroes for SkinThickness, Insulin, Glucose, BloodPressure and BMI with estimated values using KNN Imputer 
#   1C. Impute all zeroes for SkinThickness, Insulin, Glucose, Bloodpressure and BMI in dataset with median values
#   2. log transform of features with skew > 0.75 but exclude Pregnancies
#   3. Delete selected columns with low correlations coefficients
#
rem_zeroes = True
imp_zeroes = False
med_zeroes = False
log_transform = False
del_cols = True
option_string='Feature engineering steps applied: '

# Create new dataframe with copy of raw input
df = df_input.copy()


#
# 1A. Remove individuals with zero measurements for SkinThickness or Insulin
#

if rem_zeroes:
    option_string = option_string + ' Removed zero measurements '       
    in_row = df.shape[0]
    df = df.drop(df[ (df['SkinThickness'] == 0) | (df['Insulin'] == 0) ].index, inplace=False).copy()
    out_row = df.shape[0]
    print(f"\nRemoved {in_row-out_row} individuals zero measurements for SkinThickness and Insulin\n")
#
# 1B. Imputed zero values with KNN Imputer for SkinThickness, Insulin, Glucose, BloodPressure and BMI.
#

if imp_zeroes:
    option_string = option_string + ' Imputed zero values with KNNImputer '  
    df['SkinThickness'].replace(to_replace=0, value=np.nan, inplace=True)
    df['Insulin'].replace(to_replace=0, value=np.nan, inplace=True)
    df['Glucose'].replace(to_replace=0, value=np.nan, inplace=True)
    df['BloodPressure'].replace(to_replace=0, value=np.nan, inplace=True)
    df['BMI'].replace(to_replace=0, value=np.nan, inplace=True)
    
    imputer = KNNImputer(n_neighbors=5, missing_values=np.nan)
    df_imputed = pd.DataFrame(data=imputer.fit_transform(df), columns=df.columns)

    df = df_imputed.copy()
    
    print(f"\nImputed all zero measurements for SkinThickness, Insulin, Glucose, BloodPressure and BMI with KNNImputer\n")

#
# 1C. Impute zero values with Median values for SkinThickness, Insulin, Glucose, BloodPressure and BMI.
#

if med_zeroes:
    option_string = option_string + ' Imputed zero values with median value '  
    df['SkinThickness'].replace(to_replace=0, value=df['SkinThickness'].median(), inplace=True)
    df['Insulin'].replace(to_replace=0, value=df['Insulin'].median(), inplace=True)
    df['Glucose'].replace(to_replace=0, value=df['Glucose'].median(), inplace=True)
    df['BloodPressure'].replace(to_replace=0, value=df['BloodPressure'].median(), inplace=True)
    df['BMI'].replace(to_replace=0, value=df['BMI'].median(), inplace=True)
    print(f"\nImpute all zero measurements for SkinThickness, Insulin, BloodPressure and BMI in dataset with median \n")
    
#
# 2. log transform of features with skew > 0.75 but exclude Pregnacies
#

if log_transform:
    # Determine numeric columns
    num_cols = df.select_dtypes(include='number').columns

    # Calculate skew for each numeric feature and select features with skew>0.75 
    calc_skew_col = df[num_cols].skew()
    skew_cols = list(calc_skew_col[calc_skew_col > 0.75].index)

    # De-select Pregnancies and Insulin
    skew_cols.remove('Pregnancies')
    skew_cols.remove('Insulin')
    
    print(f"\nLog transforming following features: {skew_cols}\n")
    option_string = option_string + ' Log transform '
    
    for column in skew_cols:
        df.loc[:,column] = np.log(df[column]+0.00000001)
        
#
# 3. Delete selected columns with low correlations coefficients
#

if del_cols:
    # Drop columns 
    drop_cols = ['BloodPressure','SkinThickness','Insulin']    
    df.drop(columns=drop_cols, inplace=True)
    print(f"\nDropping columns {drop_cols}\n")
    option_string = option_string + ' Columns with low correlations coefficients dropped'
````

### Confusion Matrix, precision and recall

The confusion Matrix is defined as: $\left(\begin{array}{cc}True Neg & False Pos\\False Neg & True Pos\end{array}\right)$ or in abbreviated form: $\left(\begin{array}{cc}TN & FP\\FN & TP\end{array}\right)$.

Precision is defined as $\frac{(TruePos)}{(TruePos +FalsePos)}$ or $\frac{(TP)}{(TP +FP)}$

Recall is defined as $\frac{(TruePos)}{(TruePos +FalseNeg)}$ or $\frac{(TP)}{(TP +FN)}$


Precision can be used as model metric to select the best model when there is a high cost associated with a False Positive .

Recall can be used as model metric to select the best model when there is a high cost associated with a False Negative.

In code:

```code
cm=disp.confusion_matrix

TN = cm[0,0]
FP = cm[0,1]
FN = cm[1,0]
TP = cm[1,1]

precision = TP / (TP+FP)
recall    = TP / (TP+FN)
```




````python
#
# Run 6 different classifiers and print precision and recall and plot confusion matrices for each 
#

classifiers = [LogisticRegression(max_iter=5000), 
               RandomForestClassifier(criterion='entropy', random_state=0),
               GaussianNB(),
               KNeighborsClassifier(11),
               GradientBoostingClassifier(), 
               SVC(kernel='linear', C=0.01)
               ]

for cls in classifiers:
    cls.fit(X_train, y_train)

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15,15))

precision_tot = 0
recall_tot = 0

print(f'\n\033[1m    Overview of Precision and Recall for each classifier followed by confusion matrix and P-R graph for each\033[0m')
for cls, ax in zip(classifiers, axes.flatten()):
    disp=plot_confusion_matrix(cls,X_test,y_test,ax=ax,cmap='Blues',normalize=None)
    ax.title.set_text(type(cls).__name__) 
    
    cm=disp.confusion_matrix

    tn = cm[0,0]
    fp = cm[0,1]
    fn = cm[1,0]
    tp = cm[1,1]

    precision = tp / (tp+fp)
    precision_tot += precision
    recall    = tp / (tp+fn)
    recall_tot += recall

    
    print(f'    Classifier : {type(cls).__name__:<30}  Precision : {precision:<15.2f}   Recall : {recall:<15.2f}')


fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15,15))

for cls, ax in zip(classifiers, axes.flatten()):
    disp = plot_precision_recall_curve(cls, X_test, y_test, ax=ax)
    ax.set_title(f'2-class Precision-Recall curve - {type(cls).__name__}')

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15,15))

for cls, ax in zip(classifiers, axes.flatten()):
    disp = plot_roc_curve(cls, X_test, y_test, ax=ax)
    ax.set_title(f'ROC curve - {type(cls).__name__}')
    
    
print(f'\n    Average Precision : {precision_tot/6:<5.2f} Average Recall : {recall_tot/6:<5.2f}')
print(f'\n    {option_string}\n')
````

### Additional model optimisation - RandomForestClassifier hyperparameters - try to improve performance by optimising hyperparameters - Limited impact - Work in progress
1. n_estimators = number of trees in the foreset
2. max_features = max number of features considered for splitting a node
3. max_depth = max number of levels in each decision tree
4. min_samples_split = min number of data points placed in a node before the node is split
5. min_samples_leaf = min number of data points allowed in a leaf node
6. bootstrap = method for sampling data points (with or without replacement)

````python
#
# Define function to evaluate a model and return confusion matrix, precision and recall
#
def evaluate(model, X_test, y_test):
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)

    tn = cm[0,0]
    fp = cm[0,1]
    fn = cm[1,0]
    tp = cm[1,1]

    precision = tp / (tp+fp)
    recall    = tp / (tp+fn)

    return cm, precision, recall
````

````python
#
# Run 'Base' RandomForestClassifier with default parameters
#

clf_base = RandomForestClassifier(criterion='entropy', random_state=0)
      
clf_base.fit(X_train, y_train)

base_cm, base_precision, base_recall = evaluate(clf_base, X_test, y_test)
````

````python
#
# Use Random search to find hyperparameters for Grid Search
#

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]

# Number of features to consider at every split
max_features = ['auto', 'sqrt']

# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(2, 20, num = 10)]
max_depth.append(None)

# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4, 8]

# Method of selecting samples for training each tree
bootstrap = [True, False]

# Create the random grid
random_param = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}


# Initiate the RandomForestClassifier model 
clf = RandomForestClassifier(random_state=0)

# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores

random_search = RandomizedSearchCV(estimator = clf, 
                                   param_distributions = random_param, 
                                   n_iter = 100, 
                                   cv = 3, 
                                   scoring='f1_macro',
                                   verbose=2,
                                   random_state=0,
                                   n_jobs = -1)

# Fit the random search to the data
random_search.fit(X_train, y_train)
random_search.best_params_
best_random = random_search.best_estimator_

# Evaluate best model 
rand_cm, rand_precision, rand_recall = evaluate(best_random, X_test, y_test)
````

````python
#
# Create the parameter grid based on the results of random search to set up a gridsearch
#

param_grid = {
    'bootstrap': [True],
    'max_depth': [4, 6, 8],
    'max_features': ['sqrt'],
    'min_samples_leaf': [5, 8, 10],
    'min_samples_split': [1, 2, 4],
    'n_estimators': [100, 200, 400]
}

# Initiate the RandomForestClassifier model 
clf = RandomForestClassifier(random_state=0)

# Initiate the grid search model
grid_search = GridSearchCV(estimator = clf, param_grid = param_grid, cv = 3, scoring='precision_macro', n_jobs = 8, verbose = 2)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)
grid_search.best_params_
best_grid = grid_search.best_estimator_

# Evaluate best model 
grid_cm, grid_precision, grid_recall = evaluate(best_grid, X_test, y_test)
````
